```python
import numpy as np 

import pandas as pd 

import matplotlib.pyplot as plt 

import seaborn as sns 
```

```python
df = df.drop(['Status', 'unnamed1'],axis=1)
```

```python
df = df.dropna()
```

```python
df.rename(columns={'Occupation':'Sector'},inplace=True)
```

```python
df["Marital_Status"] = df["Marital_Status"].astype(str).str.replace("0","Single")
```

```python
df["Marital_Status"] = df["Marital_Status"].astype(str).str.replace("1","Married")
```

# DV

```python


c1 = ["#3886E5", "#99C2F5"]

plt.figure(figsize=(10,5))

chart_1 = sns.countplot(x="Age Group",data=df,hue="Gender",palette=c1,edgecolor="black")

for i in chart_1.containers:
    chart_1.bar_label(i)

plt.title("Age Group Wise Gender Count",family="Georgia",fontsize=16,color="#004394",fontweight="bold")

plt.xlabel("")

plt.ylabel("")

plt.show()
```

```python
C2 = [
    "#8A2D3C", "#D86572", "#FF9DA7", "#FFC9CF", "#FFEFF1"
]

D1 = df.groupby(df["Zone"],as_index=False)["Amount"].count()

plt.pie(D1["Amount"],labels=D1["Zone"],autopct="%1.f%%",startangle=105,colors=C2)

plt.title("Zone Wise Count",family="Georgia",fontsize=16,color="#7C0116",fontweight="bold")

plt.show()
```

```python
df["State"].unique()
```

```python
C3 = ["#5A2D5A", "#743B74", "#8E4D8E", "#B07AA1", "#C18EB4",
    "#D1A5C6", "#E1BDDA", "#F0D6EC", "#F8EAF7", "#FCF4FB"]

D = df.groupby("State",as_index=False)["Amount"].mean().sort_values(by="Amount",ascending=False).head(10)

plt.figure(figsize=(15,5))

sns.barplot(x="State",y="Amount",data=D,palette=C3,edgecolor="black",hue="State")

plt.title("Top 10 state Avg Amount Wise",family="Georgia",fontsize=18,color="#AD00AD",fontweight="bold")

plt.xlabel("")

plt.ylabel("")

plt.show()
```

```python

plt.figure(figsize=(15,5))

sns.lineplot(x="Sector",y="Orders",data=df , errorbar=None , marker="o",color="#B90AB1")

plt.title("Sector Wise Orders")

plt.xlabel("")

plt.ylabel("")

plt.xticks(rotation=45)

plt.grid(color="#87A9E3")

plt.show()
```
